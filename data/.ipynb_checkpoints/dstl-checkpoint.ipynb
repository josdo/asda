# dstl data pipeline

```python
from collections import defaultdict
import csv
import sys
import os
from collections import namedtuple

import cv2
from shapely.geometry import MultiPolygon, Polygon
import shapely.wkt
import shapely.affinity
import tifffile as tiff

import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from PIL import Image
```

```python
# Specify all data file paths
d = './dstl/'
dirs = {
    'raw_masks': d + 'poly/',
    'raw_imgs': d + 'tiff_imgs/',
    'in_imgs': d + 'png_imgs/',
    'in_masks': d + 'png_masks/',
    'out_imgs': d + 'imgs/',
    'out_masks': d + 'masks/',
}
```

## create masks

```python
# get image sizes
indir = dirs['raw_imgs']
dir_paths = os.listdir(indir)
tiff_paths = [f for f in dir_paths if f.endswith(".tif")]

im_size = {}
for p in tiff_paths:
    im = Image.open(indir + p)
    width, height = im.size
    _im_id = p.split('.')[0]
    im_size[_im_id] = (height, width)
```

```python
csv.field_size_limit(sys.maxsize)

# DSTL -> Potsdam label conversion map
# is this too hard-coded?
PotsdamClass = namedtuple('PotsdamClass', ['name', 'id', 'ignore_in_eval', 'color'])
classes = [
    PotsdamClass('background',  0, True, (255, 0, 0)),      # red
    PotsdamClass('building',    1, False, (0, 0, 255)),     # blue
    PotsdamClass('pavement',    2, False, (255, 255, 255)), # white
    PotsdamClass('vegetation',  3, False, (0, 255, 255)),   # turqoise
    PotsdamClass('tree',        4, False, (0, 255, 0)),     # green
    PotsdamClass('car',         5, False, (255, 255, 0)),   # yellow
]

label_map = {
    '1':  1,  # building       -> building
    '2':  0,  # misc man-made  -> background
    '3':  2,  # road           -> pavement
    '4':  0,  # track          -> background
    '5':  4,  # trees          -> tree
    '6':  3,  # crops          -> vegetation
    '7':  0,  # waterway       -> background
    '8':  0,  # standing water -> background
    '9':  5,  # vehicle large  -> car
    '10': 5,  # vehicle small  -> car
}
```

```python
# Load grid sizes
x_max = {}
y_min = {}
for _im_id, _x, _y in csv.reader(open(dirs['raw_masks'] + 'grid_sizes.csv')):
    if _x == 'Xmax' or _y == 'Ymin':
        continue    
    x_max[_im_id] = float(_x)
    y_min[_im_id] = float(_y)
```

```python
def tryParseWKT(x):
    try:
        p = shapely.wkt.loads(x)
        if type(p) == Polygon:
            p = MultiPolygon([p])
        return p
    except:
        return None

# Load train poly with shapely
train_polygons = {}
for csv_file in ['sample_submission.csv', 'train_wkt_v4.csv']:
    for _im_id, _poly_type, _poly in csv.reader(open(dirs['raw_masks'] + csv_file)):
        if _im_id == 'ImageId':
            continue
        if _im_id not in train_polygons:
            train_polygons[_im_id] = {}
        train_polygons[_im_id][_poly_type] = tryParseWKT(_poly)
```

```python
# helper function for scaling
def get_scalers(_im_id):
    h, w = im_size[_im_id]  # they are flipped so that mask_for_polygons works correctly
    w_ = w * (w / (w + 1))
    h_ = h * (h / (h + 1))
    return w_ / x_max[_im_id], h_ / y_min[_im_id]

# generate a mask given a polygons object
def mask_for_polygons(polygons, color, _im_size):
    img_mask = np.zeros(_im_size, np.uint8)
    if polygons:
        int_coords = lambda x: np.array(x).round().astype(np.int32)
        exteriors = [int_coords(poly.exterior.coords) for poly in polygons]
        interiors = [int_coords(pi.coords) for poly in polygons
                     for pi in poly.interiors]
        cv2.fillPoly(img_mask, exteriors, 1)
        cv2.fillPoly(img_mask, interiors, 0)
    
    # make channels
    img_mask = np.dstack(3 * (img_mask,)) * np.uint8(color)
    return img_mask
```

```python
outdir = dirs['in_masks']
background_color = classes[0].color

# generate and save masks
for _im_id in tqdm(train_polygons):
    x_scaler, y_scaler = get_scalers(_im_id)
    h, w = im_size[_im_id]
    mask = np.zeros((h, w, 3), np.uint8)
    for label in train_polygons[_im_id]:
        train_polygons_scaled = shapely.affinity.scale(train_polygons[_im_id][label], xfact=x_scaler, yfact=y_scaler, origin=(0, 0, 0))
        mask_color = classes[label_map[label]].color
        mask += mask_for_polygons(train_polygons_scaled, mask_color, (h, w))
    
    unmasked = np.sum(mask, axis=2) == 0
    mask += np.dstack(3 * (unmasked,)) * np.uint8(background_color)
    Image.fromarray(mask).save(outdir + _im_id + '.png', "PNG", quality=100)
```

## recolor images

```python
# Make the colors work for the tiff images
def stretch_8bit(bands, lower_percent=2, higher_percent=98):
    out = np.zeros_like(bands)
    for i in range(3):
        a = 0 
        b = 255 
        c = np.percentile(bands[:,:,i], lower_percent)
        d = np.percentile(bands[:,:,i], higher_percent)        
        t = a + (bands[:,:,i] - c) * (b - a) / (d - c)    
        t[t<a] = a
        t[t>b] = b
        out[:,:,i] =t
    return out.astype(np.uint8)
```

```python
indir = dirs['raw_imgs']
outdir = dirs['in_imgs']

dir_paths = os.listdir(indir)
tif_paths = [f for f in dir_paths if f.endswith(".tif")]

for p in tqdm(tif_paths):
    _im_id = p.split('.')[0]
    im_arr = tiff.imread(indir + p)
    im_arr = stretch_8bit(np.rollaxis(im_arr, 0, 3))
    im = Image.fromarray(im_arr)
    im.save(outdir + _im_id + '.png', 'PNG', quality=100)
```

## tiling

```python
# convert png images to tiles. 450 images, ~3000x3000   

# Splits tiff into tiles and saves each as png
def split_png(inpath, tile_pre, is_mask):
    # adapted from http://blog.odonnell.nu/posts/creating-tiles-image-python-and-pil/
    tile_ids = []
    indir = dirs['in_masks'] if is_mask else dirs['in_imgs']
    outdir = dirs['out_masks'] if is_mask else dirs['out_imgs']
    image = Image.open(indir + inpath)
    tile_width = 1000
    tile_height = 1000
    
    height, width = image.size
    left = (width % tile_width) // 2
    right = width - left - 1
    top = (height % tile_height) // 2
    bottom = height - top - 1

    currentx = left
    currenty = top
    while currenty < bottom:
        while currentx < right:
            tile = image.crop((currentx, currenty, currentx + tile_width, currenty + tile_height))
            ID = outdir + tile_pre.format(currentx, currenty)
            tile_ids.append(ID)
            tile.save(ID + ".png", "PNG", quality=100)
            currentx += tile_width
        currenty += tile_height
        currentx = left
    
    return tile_ids

partition = {
    'train': [],
    'val': [],
    'test': [],
}

for png_type in ['imgs', 'masks']:
    if png_type == 'imgs':
        outdir, indir = dirs['out_imgs'], dirs['in_imgs']
        is_img = True # flag 
    else:
        outdir, indir = dirs['out_masks'], dirs['in_masks']
        is_img = False
    
    dir_paths = os.listdir(indir)
    png_paths = [f for f in dir_paths if f.endswith(".png")]
    
    for p in tqdm(png_paths):
        #if (p != '6040_2_2.png' and p != '6040_2_2.png'):
        #    continue
        
        tile_prefix = p.split('.')[0] + '_{}_{}' # add formatters for tile row / col    
        tile_ids = split_png(p, tile_prefix, not is_img)
        
        # Partition tile ids into train/val/test folds
        if is_img:
            # shuffle ids
            rng = np.random.default_rng()
            rng.shuffle(tile_ids)
            # define partition splits
            num_tiles = len(tile_ids)
            pt1, pt2 = int(num_tiles*.7), int(num_tiles*.9)
            # assign to folds
            partition['train'] += tile_ids[:pt1]
            partition['val'] += tile_ids[pt1:pt2]
            partition['test'] += tile_ids[pt2:]

np.save('dstl_partition', partition)
```
